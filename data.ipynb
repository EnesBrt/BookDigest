```python
import json
import pandas as pd
```

```python
with open('jobs.json', 'r') as infile:
    json_data = json.load(infile)
```

```python
df = pd.json_normalize(json_data)
```

```python
df['Job Title'] = df['Job Title'].str.replace('CDI', '', case=False).str.strip()
df['Job Title'] = df['Job Title'].str.replace('cdi', '', case=False).str.strip()

df
```
